```python
from sklearn.datasets import make_classification
import numpy as np 
X,Y = make_classification(n_samples = 100, n_features = 2, n_informative = 1 , n_redundant = 0,n_classes = 2, n_clusters_per_class = 1, random_state = 42, hypercube = False, class_sep= 3)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c = Y)
```

```python
class perceptron:
    def __init__(self,X,Y,lr=0.1,epochs = 1000):
        self.X = X
        self.Y = Y
        self.lr = lr
        self.epochs = epochs
    
    def weights(self):
        X = np.insert(self.X,0,1,axis = 1)
        weight = np.ones(X.shape[1])
        for  i in range(self.epochs):
            j = np.random.randint(0,X.shape[0])
            y_hat = self.step(np.dot(X[j],weight))
            weight = weight + self.lr*(self.Y[j] - y_hat)*X[j]
        return weight[0],weight[1:]
    def step(self,z):
        return 1 if z>0 else 0
    
        
```

```python
per_1 = perceptron(X,Y)
intercept_ , coef_ = per_1.weights()
print(intercept_ , coef_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
print(f"slop: {m} and intercept : {b}")
```

```python
X_input = np.linspace(-2.5,3,1000)
y_input = m*(X_input) + b
plt.scatter(X[:,0],X[:,1],c = Y)
plt.plot(X_input,y_input)
```
