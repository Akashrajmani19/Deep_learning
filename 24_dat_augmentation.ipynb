```python
import tensorflow 
from tensorflow import keras

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
from tensorflow.keras.preprocessing import image

img_path = r'D:\gpu_libraries\deep_learning_with_tensorflow\data\image\photo-1529778873920-4da4926a72c2.jpg'

# Load the image with target size (200, 200)
img = image.load_img(img_path, target_size=(200, 200))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
datagen =ImageDataGenerator(
    rotation_range = 40, # angle
    shear_range = 0.2, # distrotion of image
    zoom_range = 0.2,
    horizontal_flip = True,
    width_shift_range = 0.2,
    height_shift_range = 0.2,fill_mode = 'reflect')
 
```

```python
img = image.img_to_array(img)
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
i = 0

for output in datagen.flow(input_batch, batch_size= 1,save_to_dir='D:\gpu_libraries\deep_learning_with_tensorflow\data'):
    i = i+1
    if i>10:
        break
```
